```python
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy import linalg

from scipy.spatial.distance import directed_hausdorff
import process_model.front_wheel_bycicle_model as pm
import config.config as config

%matplotlib qt
```

```python
process_model =pm.FrontWheelBycicleModel(vehicle_length=config.L, control_input_std=config.imu_std, dt=config.dt)
```

## Load in Simulation Data

x = x, y, v, a, theta, delta


```python

sim_data = pd.read_csv("../data/overall__imu__data.csv")
end = 100
```

```python
initial_x = [sim_data['positions_x_ground_truth'][0], sim_data['positions_y_ground_truth'][0], sim_data['velocities_ground_truth'][0], sim_data['acceleration_input'][0], sim_data['orientation_measurement'][0], sim_data['steering_input'][0]]

xs = [initial_x]
us = np.stack([sim_data['acceleration_input'], sim_data['steering_input']], axis=1)
for i in range (1,len(sim_data['timestamps'][:end])):
    xs.append(process_model.F(xs[i-1], us[i]))
    print(xs[i])

xs = np.array(xs)

```

### Positions

```python
plt.scatter(sim_data['positions_x_ground_truth'][:end], sim_data['positions_y_ground_truth'][:end], label="from sim", s=0.2)
plt.scatter(xs[:,0], xs[:,1], label="From process model", s=0.2)
plt.grid()
plt.legend()
```

### Velocities

```python
plt.plot(sim_data['timestamps'][:end], xs[:,2], label="process model")
plt.plot(sim_data['timestamps'][:end], sim_data['velocities_ground_truth'][:end], label="simulation")
plt.grid()
plt.legend()
```

### Accelerations

```python
plt.plot(sim_data['timestamps'][:end], xs[:,3], label="process model")
plt.plot(sim_data['timestamps'][:end], sim_data['acceleration_input'][:end], label="simulation")
plt.grid()
plt.legend()
```

### Steerings

```python
plt.plot(sim_data['timestamps'][:end], xs[:,5], label="process model")
plt.plot(sim_data['timestamps'][:end], sim_data['steering_input'][:end], label="simulation")
plt.grid()
plt.legend()
```
